* Sort the **data** in ascending order by **number of enrolled courses**.

```
from pyspark.sql.functions import size
```

```
users_df. \
    select('id', 'courses'). \
    withColumn('no_of_courses', size('courses')). \
    sort(size('courses')).show()
```
