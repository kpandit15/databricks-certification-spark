```
from pyspark.sql.functions import col
```

* Sort the data in ascending order by **customer_from** (default).

```
users_df. \
    select('id', 'customer_from'). \
    orderBy('customer_from'). \
    show()
```

* Sort the data in ascending order by **customer_from** with null values coming at the end.

```
users_df. \
    select('id', 'customer_from'). \
    orderBy(users_df['customer_from'].asc_nulls_last()). \
    show()
```

```
from pyspark.sql.functions import col
```

```
users_df. \
    select('id', 'customer_from'). \
    orderBy(col('customer_from').asc_nulls_last()). \
    show()
```

* Sort the data in descending order by **customer_from** (default).

```
users_df. \
    select('id', 'customer_from'). \
    orderBy(users_df['customer_from'].desc()). \
    show()
```

* Sort the data in descending order by **customer_from** with null values coming at the beginning.

```
users_df. \
    select('id', 'customer_from'). \
    orderBy(users_df['customer_from'].desc_nulls_first()). \
    show()
```
