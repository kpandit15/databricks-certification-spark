Let us make sure we understand characteristics of a Spark Data Frame.
* Spark Data Frame is nothing but a RDD with Structure
* RDD stands for Resilient Distributed Dataset. It means Reliable Distributed Collection.
* Spark Data Frames are immutable.
* As Spark Data Frames are structured, we can use structured APIs (Data Frame APIs and Spark SQL).
* Spark Data Frames are partitioned and they are distributed across executors while processing the data.
* Spark Data Frames are different from Python Pandas or R Data Frames.
