* `>`
* `<`
* `>=` (equivalent to boolean with `col1 > val1 or cal1 = val1`)
* `<=` (equivalent to boolean with `col1 < val1 or cal1 = val1`)

```
from pyspark.sql.functions import col, isnan
```

* Get Customers who paid greater than 900

```
users_df. \
    select('id', 'amount_paid'). \
    show()
```

```
users_df. \
    filter((col('amount_paid') > 900) & (isnan(col('amount_paid')) == False)). \
    select('id', 'amount_paid'). \
    show()
```

```
users_df. \
    filter('amount_paid > 900 AND isnan(amount_paid) = false'). \
    select('id', 'amount_paid'). \
    show()
```

* Get Customers who paid less than 900

```
users_df. \
    filter((col('amount_paid') < 900) & (isnan(col('amount_paid')) == False)). \
    select('id', 'amount_paid'). \
    show()
```

```
users_df. \
    filter('amount_paid < 900 AND isnan(amount_paid) = false'). \
    select('id', 'amount_paid'). \
    show()
```

* Get Customers who paid greater than or equals to 900

```
users_df. \
    filter((col('amount_paid') >= 900) & (isnan(col('amount_paid')) == False)). \
    select('id', 'amount_paid'). \
    show()
```

```
users_df. \
    filter('amount_paid >= 900 AND isnan(amount_paid) = false'). \
    select('id', 'amount_paid'). \
    show()
```

* Get Customers who paid less than or equals to 900

```
users_df. \
    filter((col('amount_paid') <= 900) & (isnan(col('amount_paid')) == False)). \
    select('id', 'amount_paid'). \
    show()
```

```
users_df. \
    filter('amount_paid <= 900 AND isnan(amount_paid) = false'). \
    select('id', 'amount_paid'). \
    show()
```

* Get the users who became customers after **2021-01-21**

```
users_df. \
    select('id', 'customer_from'). \
    show()
```

```
users_df. \
    select('id', 'customer_from'). \
    filter(col('customer_from') > '2021-01-21'). \
    show()
```

```
users_df. \
    select('id', 'customer_from'). \
    filter('customer_from > "2021-01-21"'). \
    show()
```
